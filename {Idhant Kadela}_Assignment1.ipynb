# Task 1
The coordinates can be fractional values also.

```python
import numpy as np

points = np.random.uniform(0, 100, (2,1000))

print(points)
```

## Task 2

## Part 1

```python
import matplotlib.pyplot as plt

colors = np.arange(0,100000,step=100)

x_coords = points[0]
y_coords = points[1]

plt.style.use("dark_background")
plt.scatter(x_coords, y_coords, c = colors, cmap='viridis', alpha=0.5)
plt.colorbar()
plt.show()
```
